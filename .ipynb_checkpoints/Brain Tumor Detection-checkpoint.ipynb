```python
# it needs to be read and open the image before we resize
from PIL.Image import open
df['pixels']=df['path'].map(lambda x:np.asarray(open(x).resize((224,224))))
```

```python
df_new =df[['Image','Class','path', 'pixels']]
y = np.array(df_new.Class)
```

```python
df_new.head()
```

```python
# All images pixel values are stored in array and returned shape (3762 image with size 224*224)
image_list = []
for i in range(len(df_new)):
    brain_image = df_new["pixels"][i].astype(np.float32)
    brain_image /= 255
    image_list.append(brain_image)
X = np.array(image_list)
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print('The shape of the X_train :'+' '+str(X_train.shape))
print('The size of the X_train :'+' '+str(X_train.shape[0]))
print('The shape of the X_test :'+' '+str(X_test.shape))
print('The size of the X_test:'+' '+str(X_test.shape[0]))
```

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224
EPOCHS = 50
```

```python
# Neural Networks
def model(input_shape):
    model = Sequential()
    model.add(Input(shape=input_shape))
    print(input_shape)
    
#     model.add(Conv2D(4, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
#     model.add(Conv2D(4, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
#     model.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last", padding='same'))
    
#     model.add(Conv2D(8, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
#     model.add(Conv2D(8, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
#     model.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last", padding='same'))
    
    model.add(Conv2D(16, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(Conv2D(16, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last", padding='same'))
   
            
    model.add(Conv2D(32, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(Conv2D(32, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last", padding='same'))
    
    model.add(Conv2D(64, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(Conv2D(64, kernel_size=(3,3), strides=2, padding="same", activation="relu", kernel_initializer="he_normal"))
    model.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last", padding='same'))
    

    
    model.add(Flatten())
    model.add(Dense(4, activation="relu"))
    model.add(Dense(8, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))    # Never use sigmoid for binary classification
    
    return model
```

```python
model = model(input_shape = (224, 224, 3))
```

```python
model.summary()
```

```python
optimizer = SGD(learning_rate=0.01)
# loss_fn = BinaryCrossentropy(from_logits=True)
loss_fn = BinaryCrossentropy()

model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
history = model.fit(x=X_train, y=y_train, epochs=EPOCHS, batch_size=10)
```

```python
y_pred = model.predict(X_test)
y_pred = [1 if y>=0.5 else 0 for y in y_pred]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
target_classes = ['No Tumor','Tumor']
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
```
